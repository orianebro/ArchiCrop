```python
def read_and_merge_netcdf_files(files):
    ds = xr.Dataset()
    for file in files:
        ds_read = xr.open_dataset(file)
        # Replace ids by id+len(ds)
        for i in range(len(ds_read.id.values)):
            ds_read.id.values[i] += len(ds.id.values)
        print("Merged file:", file)
    return ds
```
